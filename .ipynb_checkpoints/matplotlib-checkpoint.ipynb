## show()

One thing to be aware of: the plt.show() command should be used only once per Python session, and is most often seen at the very end of the script. Multiple show() commands can lead to unpredictable backend-dependent behavior, and should mostly be avoided.

## Two Interfaces for the Price of One

```python
# MATLAB-style interface
```

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis 
plt.subplot(2, 1, 1) # (rows, columns, panel number) 
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

Its important to note that this interface is stateful: it keeps track of the current figure
and axes

## Scatter Plots with plt.scatter

The primary difference of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

# alpha - to better see the overlapping results
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar(); # show color scale
```

```python
# x, y  - corresponds to the sepal length and width
# size - petal width, 
# color - particular species of flower
from sklearn.datasets import load_iris 
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
plt.colorbar();
```

# Continuous Errors

```python
# Gaussian process regression (GPR):
from sklearn.gaussian_process import GaussianProcess
# define the model and draw some data
model = lambda x: x * np.sin(x) 
xdata = np.array([1, 3, 5, 6, 8]) 
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1, random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True) 
dyfit = 2 * np.sqrt(MSE) # 2*sigma ~ 95% confidence region
```

```python
# Visualize the result plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                color='gray', alpha=0.2)
plt.xlim(0, 10);
```

```python
plt.plot(x, np.sin(x), '-', color='gray')
plt.fill_between(x, np.sin(x) - dy, np.sin(x) + dy, color='gray', alpha=0.2)
```

# Density and Contour Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt 
plt.style.use('seaborn-white') 
import numpy as np
```

## Visualizing a Three-Dimensional Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

# !!!
X, Y = np.meshgrid(x, y) 

Z = f(X,Y)

# RdGy (short for Red-Gray)
plt.contour(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```
